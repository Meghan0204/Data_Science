```python
# How many of each payment type are represented in the data?

df['payment_type'].value_counts()

# Payment types
# 1= Credit card 
# 2= Cash 
# 3= No charge 
# 4= Dispute 
# 5= Unknown 
# 6= Voided trip
```

# According to the data dictionary, the payment method was encoded as follows:

1 = Credit card  
2 = Cash  
3 = No charge  
4 = Dispute

6 = Voided trip

```python
# What is the average tip for trips paid for with credit card?

credt_card_transactions = df
credt_card_transactions = credt_card_transactions[credt_card_transactions['payment_type']==1]
credt_card_transactions['tip_amount'].mean()
```

```python
# What is the average tip for trips paid for with cash?
cash_transactions = df
cash_transactions = cash_transactions[cash_transactions['payment_type']==2]
cash_transactions['tip_amount'].mean()
```

```python
# Do any fairs paid in cash have tips that were reported?  No
cash_transactions = df
cash_transactions = cash_transactions[cash_transactions['payment_type']==2]
cash_transactions['tip_amount'].max()
```

```python
# How many times is each vendor ID represented in the data?
df['VendorID'].value_counts()
```

```python
# What is the mean total amount for each vendor?
df.groupby('VendorID')['total_amount'].mean()
```

```python
# Filter the data for credit card payments only

credt_card_transactions = df[df['payment_type']==1]
credt_card_transactions.head(2)
```

```python
# Calculate the average tip amount for each passenger count (credit card payments only)
credt_card_transactions.groupby('passenger_count')['tip_amount'].mean()
```

```python
df.groupby('VendorID')['total_amount'].min()
```

```python
vendor1 = df
vendor1 = vendor1[vendor1['VendorID']==1]
vendor1.describe()
```

```python
# Correlation values run from –1 to +1:
# ±0.00 – ±0.19 → very weak (basically none)
# ±0.20 – ±0.39 → weak
# ±0.40 – ±0.59 → moderate
# ±0.60 – ±0.79 → strong
# ±0.80 – ±1.00 → very strong

plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```
